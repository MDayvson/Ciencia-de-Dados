### Avaliando o erro na regressão

```python
import pandas as pd
dados = {
    'X': [1,2,3,4],
    'Y': [5,7,10,11]
}

dados = pd.DataFrame(dados)

X = dados.X.values.reshape(-1,1)
y = dados.Y
```

```python
dados.shape[0] # Número de linhas da base
```

```python
X.shape[1] # Número de colunas do modelo
```

```python
# Regressão Linear
from sklearn import linear_model
regLinear = linear_model.LinearRegression()
regLinear.fit(X,y)
```

```python
# Visualizando essa reta
a=regLinear.coef_[0]
b=regLinear.intercept_
y_reta = a*X+b

import matplotlib.pyplot as plt
fig,ax = plt.subplots()

ax.scatter(X,y)
ax.plot(X,y_reta,'--r')

plt.show()
```

```python
y_pred_reg = regLinear.predict(X)
```

```python
# Árvore de Decisão
from sklearn.tree import DecisionTreeRegressor
regArvore = DecisionTreeRegressor(random_state=0)
regArvore.fit(X,y)
```

```python
from sklearn import tree
tree.plot_tree(regArvore);
```

```python
y_pred_arvore = regArvore.predict(X)
```

#### O coeficiente de determinação (R quadrado)
- https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score

```python
from sklearn.metrics import r2_score
```

```python
y_true = y
```

```python
# Regressão Linear
r2_score(y_true, y_pred_reg)
```

```python
# Árvore de Decisão
r2_score(y_true, y_pred_arvore)
```

#### O coeficiente de determinação ajustado

```python
# Regressão Linear
r2_score_modelo = r2_score(y_true, y_pred_reg)
somatorio = 1 - r2_score_modelo
n = dados.shape[0]
p = X.shape[1] 


r2_ajus = 1 - somatorio*((n-1)/(n-p-1))
print(r2_ajus)
```

```python
# Árvore de Decisão
r2_score_modelo = r2_score(y_true, y_pred_arvore)
somatorio = 1 - r2_score_modelo
n = dados.shape[0]
p = X.shape[1] 


r2_ajus = 1 - somatorio*((n-1)/(n-p-1))
print(r2_ajus)
```
