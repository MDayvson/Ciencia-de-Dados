### Comandos básicos do SQL

```python
# Importando o pandas
import pandas as pd
```

```python
# Criando a conexão
import sqlite3
con = sqlite3.connect('BaseDados.db')
```

```python
# Criando um cursor
cur = con.cursor()
```

```python
# Criando uma função para consultar os dados
def executa_sql(comando):
    cur.execute(comando)
    resultado = cur.fetchall()
    resultado = pd.DataFrame(resultado)
    resultado.columns = [i[0] for i in cur.description]
    print(resultado.shape)
    display(resultado.head())
    return resultado
```

**Quais alunos estão com problema de acesso?**
- Para fazer um filtro na base, podemos utilizar o `WHERE`

```python
# Filtrando apenas alunos com problema de acesso usando o pandas
resultado = executa_sql('SELECT * FROM dados')
```

```python
resultado = resultado[resultado.acesso_plataforma == 0]
resultado.shape
```

```python
# Fazendo esse mesmo filtro utilizando o WHERE do SQL
resultado = executa_sql('SELECT * FROM dados WHERE acesso_plataforma = 0')
```

**Existe algum aluno que está com problema de acesso E que entrou na plataforma nos últimos 2 dias? E algum aluno com problema OU que não entrou na plataforma nos últimos 10 dias?**
- Para colocar mais de 1 condição no `WHERE`, podemos usar o `AND` ou o `OR`
- Para negar uma condição, podemos utilizar o `NOT`

```python
# Verificando se existe algum aluno com problema de acesso E que entrou nos últimos 2 dias
resultado = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula, [e-mail] as email FROM dados \
                        WHERE acesso_plataforma = 0 AND dias_ultimo_acesso <= 2')
```

```python
# Verificando problema de acesso OU acesso acima de 10 dias
resultado = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula, [e-mail]  as email, \
                        acesso_plataforma, dias_ultimo_acesso FROM dados \
                        WHERE acesso_plataforma = 0 OR dias_ultimo_acesso > 10')
```

```python
display(resultado)
```

```python
# Buscando alunos com mais de 12 dias de acesso
resultado = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula, [e-mail]  as email, \
                        acesso_plataforma, dias_ultimo_acesso FROM dados \
                        WHERE dias_ultimo_acesso >= 12')
```

```python
# Negando essa condição acima com NOT
resultado = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula, [e-mail]  as email, \
                        acesso_plataforma, dias_ultimo_acesso FROM dados \
                        WHERE NOT dias_ultimo_acesso >= 12')
```

```python
resultado = executa_sql('SELECT DISTINCT nome_aluno, cod_matricula, [e-mail]  as email, \
                        acesso_plataforma, dias_ultimo_acesso FROM dados \
                        WHERE dias_ultimo_acesso < 10')
```
