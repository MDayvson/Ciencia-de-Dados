- **Criando os pontos**

```python
# Importando o dataset iris
from sklearn.datasets import load_iris
X,y = load_iris(return_X_y=True,as_frame=True)

# Selecionando apenas as colunas de pétala
X1 = X.loc[:,['petal length (cm)','petal width (cm)']]
```

- **Utilizando o K-Means**

```python
# Importando o KMeans
from sklearn.cluster import KMeans

# Utilizando o algoritmo
kmeans2 = KMeans(n_clusters=2, random_state=0).fit(X1)

# Definindo o ponto que estamos verificando
ponto = X1[X1['petal length (cm)'] == 3]
```

```python
# Visualizando graficamente
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.scatter(X1['petal length (cm)'],X1['petal width (cm)'],c=kmeans2.labels_)
ax.scatter(ponto['petal length (cm)'],ponto['petal width (cm)'],c='b',s=50)
ax.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],c='r')

plt.show()
```

- **Podemos traçar os triângulos vistos acima**

```python
x_roxo = kmeans2.cluster_centers_[0][0]
y_roxo = kmeans2.cluster_centers_[0][1]
```

```python
x_ponto = ponto['petal length (cm)']
y_ponto = ponto['petal width (cm)']
```

```python
# Traçando a reta indo do x do cluster roxo até o x do ponto azul
fig, ax = plt.subplots()

ax.scatter(X1['petal length (cm)'],X1['petal width (cm)'],c=kmeans2.labels_)
ax.scatter(ponto['petal length (cm)'],ponto['petal width (cm)'],c='b',s=50)
ax.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],c='r')
ax.plot([x_roxo,x_ponto],[y_roxo,y_roxo],'--r')

plt.show()
```

```python
# Calculando essa distância
x_ponto-x_roxo
```

```python
# Traçando a reta indo do y do cluster até o y do ponto azul
fig, ax = plt.subplots()

ax.scatter(X1['petal length (cm)'],X1['petal width (cm)'],c=kmeans2.labels_)
ax.scatter(ponto['petal length (cm)'],ponto['petal width (cm)'],c='b',s=50)
ax.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],c='r')
ax.plot([x_roxo,x_ponto],[y_roxo,y_roxo],'--r')
ax.plot([x_ponto,x_ponto],[y_roxo,y_ponto],'--r')

plt.show()
```

```python
# Calculando essa distância
y_ponto-y_roxo
```

```python
# Traçando a distância euclidiana desse ponto
fig, ax = plt.subplots()

ax.scatter(X1['petal length (cm)'],X1['petal width (cm)'],c=kmeans2.labels_)
ax.scatter(ponto['petal length (cm)'],ponto['petal width (cm)'],c='b',s=50)
ax.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],c='r')
ax.plot([x_roxo,x_ponto],[y_roxo,y_roxo],'--r')
ax.plot([x_ponto,x_ponto],[y_roxo,y_ponto],'--r')
ax.plot([x_roxo,x_ponto],[y_roxo,y_ponto],'--g')

plt.show()
```

```python
# Fazendo o cálculo dessa distância
((x_roxo-x_ponto)**2+(y_roxo-y_ponto)**2)**(1/2)
```
