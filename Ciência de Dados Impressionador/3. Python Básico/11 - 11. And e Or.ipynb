# Casos com várias condições/comparações

### Estrutura:

Quando temos várias comparações, ao invés de criar if dentro de if podemos usar os operadores "and" e "or" para tratar essas condições.

Funciona assim:

### Exemplo

Vamos voltar ao exemplo de cálculo de meta de vendas dos funcionários. Muitas empresas atribuem bonificação do salário dos funcionários de acordo com o resultado do funcionário e também com o resultado da empresa como um todo.

Nesse caso, a regra funciona da seguinte forma:
- Se o funcionário vendeu mais do que a meta de vendas e a loja bateu a meta de vendas da loja, o funcionário ganha 3% do que ele vendeu em forma de bônus.
- Caso o funcionário tenha batido a meta de vendas individual dele, mas a loja não tenha batido a meta de vendas da loja como um todo, o funcionário não ganha bônus.

```python
def bonus_funcionario(meta_funcionario, meta_loja, vendas_funcionario, vendas_loja):
    if vendas_funcionario >= meta_funcionario and vendas_loja >= meta_loja:
        bonus = vendas_funcionario * 0.03
        percentual = '3%'
        msn = 'Funcionário e loja atingiram as metas. Bônus de 3% do valor da venda.'
    elif vendas_funcionario >= meta_funcionario or vendas_loja >= meta_loja:
        bonus = 0
        percentual = '0%'
        if vendas_funcionario >= meta_funcionario:
            msn = 'Funcionário atingiu a meta, mas a loja não. Sem bônus.'
        elif vendas_loja >= meta_loja:
            msn = 'Loja atingiu a meta, mas o funcionário não. Sem bônus.'
    else:
        msn = 'Funcionário e loja não atingiram as metas. Sem bônus.'

    return bonus, percentual, msn
```

```python
# dados do exemplo:
meta_f, meta_l = 10000, 50000
vendas_f, vendas_l = 11000, 80000
```

```python
bonus_final, taxa_bonus, mensagem = bonus_funcionario(meta_f, meta_l, vendas_f, vendas_l)
print(mensagem)
print(f'Bônus do funcionário: R${bonus_final:.2f} ({taxa_bonus} do valor da venda)')
```

### Outro exemplo

Agora vamos levar essa análise mais a fundo.

Nessa empresa, existe um outro caso também que garante que o funcionário ganhe um bônus, independente das vendas que ele fez naquele mês.

Todo mês os diretores da empresa fazem uma avaliação qualitativa de todos os funcionários. Nessa avaliação os diretores dão uma nota de 0 a 10 para cada funcionário. Se a nota do funcionário for 9 ou 10, ele também ganha o bônus de 3% do valor de vendas. (os bônus não são cumulativos)

```python
# condição 2: teve avaliação de desempenho com nota 9 ou 10.
def calcular_bonus_final(meta_funcionario, meta_loja, vendas_funcionario, vendas_loja, nota_desempenho):
    if (vendas_funcionario >= meta_funcionario and vendas_loja >= meta_loja) or nota_desempenho >= 9:
        bonus = vendas_funcionario * 0.03
        percentual = '3%'
        msn = 'Funcionário e loja atingiram as metas ou o funcionário teve uma ótima avaliação de desempenho. Bônus de 3% do valor da venda.'
    elif vendas_funcionario >= meta_funcionario or vendas_loja >= meta_loja:
        bonus = 0
        percentual = '0%'
        if vendas_funcionario >= meta_funcionario:
            msn = 'Funcionário atingiu a meta, mas a loja não. Sem bônus.'
        elif vendas_loja >= meta_loja:
            msn = 'Loja atingiu a meta, mas o funcionário não. Sem bônus.'
    else:
        bonus = 0
        percentual = '0%'
        msn = 'Funcionário e loja não atingiram as metas e o funcionário não teve uma ótima avaliação de desempenho. Sem bônus.'
    
    return bonus, percentual, msn
```

```python
# Teste
nota_do_funcionario = 8
meta_f, meta_l = 10000, 50000
vendas_f, vendas_l = 8000, 40000 # Não bateu as metas de venda
```

```python
bonus_final, taxa_bonus, mensagem = calcular_bonus_final(meta_f, meta_l, vendas_f, vendas_l, nota_do_funcionario)

print(mensagem)
print(f'Bônus do funcionário: R${bonus_final:.2f} ({taxa_bonus} do valor da venda)')
```
